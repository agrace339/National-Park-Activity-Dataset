```python
# import libraries
import pandas as pd
import numpy as np
import ast
```

```python
path = "Data/Park Activities.csv"

df = pd.read_csv(path)
df.head()
```

```python
# This code prints the dataset’s shape, column data types, missing value counts, and column names.

print("Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isna().sum())
df.columns
```

```python
# standardizing of column names by making them lowercase and replacing spaces or hyphens with underscores.
df.columns = (
    df.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
    .str.strip()
)
df.columns
```

```python
# removes any duplicate rows from the dataset
df.drop_duplicates(inplace=True)
```

```python
# This code fills missing latitude and longitude values with NaN to keep coordinates consistent and avoid incorrect replacements.
df = df.fillna({
    "latitude": np.nan,
    "longitude": np.nan,
})
```

```python
# This code cleans text fields by converting park names to title case and removing extra whitespace from both name and description.

df["park_name"] = df["park_name"].str.title().str.strip()
df["park_description"] = df["park_description"].str.strip()
```

```python
# This code identifies and displays any rows containing invalid latitude or longitude values outside the real-world geographic ranges.

invalid_coords = df[(df["latitude"] > 90) | (df["latitude"] < -90) |
                    (df["longitude"] > 180) | (df["longitude"] < -180)]
invalid_coords
```

```python
df.rename(columns={'unnamed:_0': 'SNo'}, inplace=True)
```

```python
# This code checks how many unique park codes exist and compares it to the total number of rows in the dataset.
df["parkcode"].nunique(), df.shape[0]
```

```python
# This code removes HTML tags from all text-based columns by applying a BeautifulSoup cleaning function to every object-type column.
from bs4 import BeautifulSoup, MarkupResemblesLocatorWarning
import pandas as pd
import warnings

warnings.filterwarnings("ignore", category=MarkupResemblesLocatorWarning)

def remove_html_tags(text):
    if pd.isna(text):
        return ""

    if not isinstance(text, str):
        return text
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text(separator=" ").strip()

for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].apply(remove_html_tags)
```

```python
# This code fixes encoded characters in the *longdescription* column by converting HTML escape sequences back into readable text.
import html

df["longdescription"] = df["longdescription"].apply(lambda x: html.unescape(str(x)))
```

```python
# Replace multiple whitespace characters in 'longdescription' with a single space
df["longdescription"] = df["longdescription"].str.replace(r"\s+", " ", regex=True)
```

```python
# Fix text encoding issues (mojibake) in all string/object columns of the DataFrame
def fix_mojibake(text):
    try:
        return text.encode("latin1").decode("utf8")
    except:
        return text

for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].astype(str).apply(fix_mojibake)
```
